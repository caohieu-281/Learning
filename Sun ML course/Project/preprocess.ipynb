```python
import pandas as pd
import numpy as np
import json
```

```python
df_train = pd.read_csv('train.tsv',delimiter='\t',encoding='utf-8')
df_train
```

```python
boilerplate = df_train['boilerplate']

dicts = {
    'title' : [],
    'body' : [],
    'url' : []
}
for i in range(len(boilerplate)):
    temp = json.loads(boilerplate[i])
    for j in dicts.keys():
        dicts[j].append(temp.get(j, None))
```

```python
dicts['url_boilerplate'] = dicts.pop('url')
for i in dicts.keys():
    df_train[i] = dicts[i]
```

```python
df_train.to_csv('new_train_data.tsv', index=False)
```
