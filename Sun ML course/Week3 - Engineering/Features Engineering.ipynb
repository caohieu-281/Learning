```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
dataset = load_boston()
```

```python
X, y = dataset.data, dataset.target
```

```python
X.shape, y.shape
```

```python
X = X[:, -1]
```

```python
plt.scatter(X, y)
```

```python
a = b = 0
```

```python
def plot():
    plt.clf()
    plt.scatter(X, y)
    x_range = np.array([np.min(X), np.max(X)])
    plt.plot(x_range, a * x_range + b, color= 'violet', linewidth= 5)
    plt.show()
```

```python
plot()
```

```python
X_normalize = ((X - np.min(X)) / (np.max(X) - np.min(X)))
y_normalize = ((y - np.min(y)) / (np.max(y) - np.min(y)))
```

```python
def plot_nor():
    plt.clf()
    plt.scatter(X_normalize, y_normalize)
    x_range = np.array([np.min(X_normalize), np.max(X_normalize)])
    plt.plot(x_range, a * x_range + b, color= 'violet', linewidth= 5)
    plt.show()
```

```python
learning_rate = 1e-2
```

```python
a = b = 0
```

```python
for _ in range(10000):
    d_a = -np.sum((y_normalize - a * X_normalize - b) * X_normalize) / X_normalize.size
    d_b = -np.sum(y_normalize - a * X_normalize - b) / X_normalize.size

    a -= learning_rate * d_a
    b -= learning_rate * d_b

plot_nor()
```
